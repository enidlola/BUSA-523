### Histogram

```python
#A. Histogram

#Histograms visualize quantitative data or numerical data, whereas bar charts display categorical variables.

# importing libraries

import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd

# load the Data.csv dataset

data = pd.read_csv("data.csv")


# view the dataset

#print(data.head(14))
 


plt.hist(data['dept_1 sales'],

         alpha=0.5,

         label='Dept1 Sales',

         color='red') # customized color parameter


plt.title("Dept 1 Sales by Day Type")

plt.show()
```

```python
#Histogram - plotting overlapping histograms for 3 department sales


import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd

# load the Data.csv dataset

data = pd.read_csv("data.csv")


# view the dataset

#print(data.head(14))
 


plt.hist(data['dept_1 sales'],

         alpha=0.5,

         label='Dept1 Sales',

         color='red') # customized color parameter

plt.hist(data['dept_2 sales'],

         alpha=0.5,

         label='Dept2 Sales',

         color='green')

 

plt.hist(data['dept_3 sales'],

         alpha=0.5,

         label='Dept3 Sales',

         color='yellow')

plt.legend(loc='upper right')

plt.title("Dept 1, 2 & 3 Sales Histogram")

plt.show()
```

### Bar Chart

```python
#B. Bar chart  

#bar charts display categorical variables

#Unlike histograms, bar charts have a finite set of categories

#Bar Chart on the effect of 'Day Type' on Department 1 Sales.

import matplotlib.pyplot as plt

import pandas as pd

import matplotlib as mlp

df = pd.read_csv('data.csv', index_col = 9)

#df -DataFrame

df1 = df.sort_values('Day type', ascending=False)


X = list(df1.iloc[:, 9])    #List of Day Type

Y = list(df1.iloc[:, 10])   #List of Dept 1 Sales

 

# Plot the data using bar() method

plt.bar(X, Y, color='g')

plt.title("Dept 1 Sales by Day Type")

plt.xlabel("Day Type")

plt.ylabel("Dept Sales")

plt.show()
```

```python
#Bar Chart on the Department 1 Sales by year.

import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv('data.csv')

#Create and Add another column that extracts the year from Date Column

df['Year'] = df['Date'].str[-4:]

#Replaceing spaces in column names with _

df.columns = [c.replace(' ', '_') for c in df.columns]

#Sort the Year column in ascending order

df1 = df.sort_values('Year', ascending=True)

#Plot the data using bar() for the Bar Chart - X = Year and Y = Sales 1 Dept Data.

plt.bar(df1.Year, df1.dept_1_sales, color='y')

plt.title("Dept Sales by Year")

plt.xlabel("Years")

plt.ylabel("Dept Sales")
```

```python
#Bar Chart on the 'Department 1, 2, 3 Sales by year.

import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv('data.csv')

#Create and Add another column that extracts the year from Date Column

df['Year'] = df['Date'].str[-4:]

df.columns = [c.replace(' ', '_') for c in df.columns]

#Sort the Year column in ascending order

df1 = df.sort_values('Year', ascending=True)

#Plot the data using bar() for the Bar Chart - X = Year and Y = Sales (1/2/3) Dept Data.

plt.bar(df1.Year, df1.dept_1_sales, color='g')

plt.bar(df1.Year, df1.dept_2_sales, color='r')

plt.bar(df1.Year, df1.dept_3_sales, color='b')

plt.title("Dept 1,2 3 Sales by Year")

plt.xlabel("Years")

plt.ylabel("Dept Sales")
```

```python
#Line Chart on the effect of 'Temperature' on three Department Sales.

#Names of Columns used for the DataSet: Temperature, dept_1 sales, dept_2 sales, dept_3 sales

import matplotlib.pyplot as plt

import pandas as pd

import matplotlib as mlp

from matplotlib.pyplot import figure

df = pd.read_csv('data.csv')

#What are the columns in this dataframe:

#df.columns

#Removing spaces in column name with _:

df.columns = [c.replace(' ', '_') for c in df.columns]

df1 = df.sort_values('Temperature', ascending=True)

#Figure Size

figure(figsize=(10, 8), dpi=80)

#Ploting

plt.plot(df1.Temperature, df1.dept_1_sales)

plt.plot(df1.Temperature, df1.dept_2_sales)

plt.plot(df1.Temperature, df1.dept_3_sales)

plt.title("Three Departmental Sales on Temperature")

plt.xlabel("Temperature")

plt.ylabel("Dept 1, 2, 3 Sales")

plt.legend(["Dept 1 Sales", "Dept 2 Sales", "Dept 3 Sales"])

plt.show()
```

# Area plot

```python
#D. Area chart

# Using CPI for y-axis

# What is the impact on the bottom-line (Sales) for 3 depts?

#Import the Libraries

import matplotlib.pyplot as plt

%matplotlib inline

from matplotlib.pyplot import figure

df = pd.read_csv('data.csv')

#Remove spaces in column name with _:

df.columns = [c.replace(' ', '_') for c in df.columns]

#Plotting begins

#Figure Size

figure(figsize=(12, 10), dpi=80)

plt.title("Three Departmental Sales on Consumer Price Index (CPI)")

plt.xlabel("Consumer Price Index (CPI)")

plt.ylabel("Dept 1, 2, 3 Sales")

plt.legend(["Dept 1 Sales", "Dept 2 Sales", "Dept 3 Sales"])

plt.stackplot(df.CPI,df.dept_1_sales, df.dept_2_sales, df.dept_3_sales)

plt.show()
```
