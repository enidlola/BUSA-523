### A Coin Flipping Program- List

 - Flip a coin 10,000 times
 - Sore the results in a list
 - Identify the number of streaks (streak: a series of 5 or more heads or tails in a row)

```python
"""
Program: A Coin Flipping Game 
Project: Assignment 2 - Question 1
Professor: Dr. Vinayaka Gude
Student: Eniola Ayanlola
"""
#************************************************************************************************************

# import the random modules

import random

myList = []

n = int(input('Type in how many times you would like to toss the coin : '))

i = 0

while i < n:                       #Create a list containing (H)eads and (T)ails

    outcome = random.randint(0,1)   #Generate a random number 0 (head) or 1 (tail)

    if outcome == 0 :

        toss = 'H'

        myList.append(toss)

    elif outcome == 1 :

        toss = 'T'

        myList.append(toss)

    i += 1

# Now let's sort through the list for (H)eads

headStreak = 0

total_H_Streak = 0



for i in myList:

    if i == 'H' :

        headStreak += 1

        if headStreak == 5:

            total_H_Streak += 1

    else :

        headStreak = 0      #reset it

print('The streak of 5 or more heads in a row is : ',total_H_Streak)

# Now let's sort through the list for (T)ails

tailStreak = 0

total_T_Streak = 0

for j in myList :

    if j == 'T' :

        tailStreak += 1

        if tailStreak == 5:

        
            total_T_Streak += 1

    else :

        tailStreak = 0

print('The Streak of 5 or more tails in a row is : ',total_T_Streak)
```

### A 3 player Die Summation Program


- There are 3 players and 10 iterations.
-  In each iteration, every player rolls a die.
-  The winner of the game is the one who has the highest score when rolls from all the iterations are added.

```python
"""
Program: A 3 player dice summation Program
Project: Assignment 2 Question 4
Professor: Dr. Vinayaka Gude
Student: Eniola Ayanlola
"""
#*******************************************
# import the random module
import random
print("This is a die game for 3 players that roll a die 10 times. The player with the highest score is the Winner.")
dummyV = input("Press Enter to Play!!!!!!!!")
n = 10      #Number of iterations - 10
tPlayer1 = 0
tPlayer2 = 0
tPlayer3 = 0
i = 0
while i <= n:       #10 iterations are allowed
    outcome1 = random.randint(1,6)   #Outcome of the 1st player
    outcome2 = random.randint(1,6)   #Outcome of the 2nd player
    outcome3 = random.randint(1,6)   #Outcome of the 3rd player
    #Accumulate Totals for all the players
    tPlayer1 += outcome1
    tPlayer2 += outcome2
    tPlayer3 += outcome3
    i = i + 1
# Exit the loop to determine has the highest score.
if (tPlayer1 > tPlayer2) and (tPlayer1 > tPlayer3):
    print("Player 1 is the Winner! with a score of: ",tPlayer1)
    print("Player 2 score is:", tPlayer1, "and Player 3 is: ", tPlayer3)
elif (tPlayer2 > tPlayer1) and (tPlayer2 > tPlayer3):
    print("Player 2 is the Winner! with a score of: ",tPlayer2)
    print("Player 1 score is:", tPlayer1, "and Player 3 is: ", tPlayer3)
elif (tPlayer3 > tPlayer1) and (tPlayer3 > tPlayer2):
    print("Player 3 is the Winner! with a score of: ",tPlayer3)
    print("Player 1 score is:", tPlayer1, "and Player 2 is: ", tPlayer2)
else:
    print("There is no Winner because scores for Player 1, Player 2, and Player 3 are: tPlayer1, tPlayer2, tPlayer3")
```
